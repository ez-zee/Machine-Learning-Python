```python
import numpy as np

# Данные: площадь (x) и цена (y)
X = np.array([100, 150, 200])
y = np.array([250, 350, 450])

# Добавляем столбец единиц для смещения
X_with_bias = np.c_[X, np.ones(len(X))]

# Находим веса и смещение методом наименьших квадратов
coefficients = np.linalg.inv(X_with_bias.T @ X_with_bias) @ X_with_bias.T @ y
w, b = coefficients

print(f"Веса: w = {w:.2f}, смещение b = {b:.2f}")
# Вывод: w = 2.00, b = 50.00

# Проверка предсказания
x_test = 120
y_pred = w * x_test + b
print(f"Предсказанная цена для площади {x_test} м²: {y_pred:.2f} тыс. $")  # 290.00 тыс. $
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Данные: площадь (x) и цена (y)
X = np.array([100, 150, 200])
y = np.array([250, 350, 450])

# Добавляем столбец единиц для смещения
X_with_bias = np.c_[X, np.ones(len(X))]

# Находим веса и смещение методом наименьших квадратов
coefficients = np.linalg.inv(X_with_bias.T @ X_with_bias) @ X_with_bias.T @ y
w, b = coefficients

# Создаем график
plt.scatter(X, y, color='blue', label='Данные')  # Отображаем исходные данные
plt.plot(X, w * X + b, color='red', label='Линия регрессии')  # Отображаем линию регрессии

# Настраиваем график
plt.xlabel('Площадь (м²)')
plt.ylabel('Цена (тыс. $)')
plt.title('Линейная регрессия')
plt.legend()
plt.grid(True)

# Отображаем график
plt.show()
```
